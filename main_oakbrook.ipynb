```python

# df = pd.read_csv('/Users/shivangsingh/Downloads/20250115 - Ares.csv')
df_oci.columns = df_oci.columns.str.strip()

book_size_ares = df_oci.loc[df_oci["YearMonthStatus"] == "OPEN", "PrincipalBalance"].sum()
print("book_size", book_size_ares)
# WA APR 
wa_apr_numerator = (df_oci["PrincipalBalance"] * df_oci["APR"] * (df_oci["YearMonthStatus"] == "OPEN")).sum()
wa_apr_denominator_a = df_oci.loc[df_oci["YearMonthStatus"] == "OPEN", "PrincipalBalance"].sum()
wa_apr_ares = wa_apr_numerator / wa_apr_denominator_a if wa_apr_denominator_a != 0 else 0
print("wa_apr", wa_apr_ares)
month_value = df_oci["MonthOfRecruitment"].iloc[-1]
total_originations_ares = df_oci.loc[
    (df_oci["YearMonthStatus"] == "OPEN") & (df_oci["MonthOfRecruitment"] == month_value),
    "PrincipalBalance"
].sum()
print("total_originations", total_originations_ares)
```
